```python
## Given the system of inequalities below, determine the shape of the feasible region and find the vertices of the feasible region. Give the shape as "triangle", "quadrilateral", or "unbounded". Report your vertices starting with the one which has the smallest x-value. If more than one vertex has the same, smallest x-value, start with the one that has the smallest y-value. Proceed clockwise from the first vertex.

x+y≥77x+y≥12x≥0y≥0
```

```python
import numpy as np, matplotlib.pyplot as plt
from scipy.optimize import linprog

print("Maximize 3x+2y with constraints x+y>=7, 7x+y>=12, x,y>=0")

corners = [(0,12),(7,0),(5/6,37/6)]
vals = [3*x+2*y for x,y in corners]
best = corners[np.argmax(vals)]
print("\nCorners:", corners)
print("3x+2y values:", vals)
print("Best corners =", best, "Value =", max(vals))

x = np.linspace(0,12,400)
plt.plot(x,7-x,label="x+y=7")
plt.plot(x,12-7*x,label="7x+y=12")
plt.fill_between(x,np.maximum(7-x,12-7*x),14,color="lightblue",alpha=.4)
plt.scatter(*zip(*corners),color="k")
plt.show()

print("\nSlack variables (s1=x+y-7, s2=7x+y-12):")
for x,y in corners: print(f"x={x:.2f}, y={y:.2f}, s1={x+y-7:.2f}, s2={7*x+y-12:.2f}")

A,b=[[-1,-1],[-7,-1]],[-7,-12]
res=linprog(c=[-3,-2],A_ub=A,b_ub=b,bounds=[(0,None),(0,None)],method='highs')
print("\nSolver:", res.x, "Max 3x+2y =", -res.fun)
```

##  A company manufactures two models of snowboards, standard and deluxe. Each deluxe model requires 30 hours to produce, and 66 units of material. Each standard model requires 15 hours to produce and 55 units of material. The company has 960 production hours available and 2596 units of material in stock. The deluxe model sells for $320 and the standard model sells for $200.

## What amount of each model should be produced and sold to obtain the maximum revenue?

```python
import numpy as np, matplotlib.pyplot as plt
from scipy.optimize import linprog

print(r"Max R=320x+200y; with constraints 30x+15y≤960, 66x+55y≤2596, x,y≥0")

pts=[(0,47.2),(32,0),(21,22)]
vals=[320*x+200*y for x,y in pts]
print("Corners:",pts,"\nMax corner:",pts[np.argmax(vals)],"R=",max(vals))


x=np.linspace(0,35)
plt.plot(x,(960-30*x)/15)
plt.plot(x,(2596-66*x)/55)
plt.fill_between(x,0,np.minimum((960-30*x)/15,(2596-66*x)/55),color='lightblue')
plt.scatter(*zip(*pts));
plt.show()

for x,y in pts: print(f"x={x},y={y},s1={960-(30*x+15*y)},s2={2596-(66*x+55*y)}")

A=[[30,15],[66,55]];b=[960,2596]
r=linprog(c=[-320,-200],A_ub=A,b_ub=b)
print("Solved = ",r.x,"Max = ",-r.fun)
```
